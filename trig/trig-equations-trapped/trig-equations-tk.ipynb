```python
def solution_point_and_label(angle): #angle in degrees
    if (angle % 40 == 0): 
        color = 'red'
        horizontal_alignment = 'left'
        rot = -45
    else: 
        color = 'blue'
        horizontal_alignment = 'right'
        rot = -45
    solution_point =  point( (angle*pi/180, sqrt(3) / 2 ), color =  color, pointsize=30, zorder= 100)
    solution_label = text("$" + str(angle) + "^\circ$", ( angle*pi/180,  sqrt(3)/2), 
    rotation = rot,
    color = color,
    horizontal_alignment = horizontal_alignment, 
    vertical_alignment='bottom')
    return solution_point + solution_label

    
```

```python


f(theta) = (sin(theta))^2 - 1/4

xt = [k * pi for k in range(-2,4)]
xt_lbl = ['$-2\pi$', '$-\pi$', '', '$\pi$', '$2\pi$', '$3\pi$',]

solutions = [-13*pi/6, -11*pi/6, -7*pi/6, -5*pi/6, -pi/6, pi/6, 5*pi/6, 7*pi/6, 11*pi/6, 13*pi/6]

solution_points = points([(theta, f(theta)) for theta in solutions ], 
                                                 color = 'blue', 
                                         pointsize=30,
                                        zorder= 100)

trig_plot = plot(f,
         ( theta,-2.5*pi,2.5*pi),
             ticks = [xt, [-1,+1]],
             tick_formatter = [xt_lbl, ['$-1$','$+1$']],
     thickness = 3,
     color = 'red',
     aspect_ratio = 1,
    ymin = -1.2,
    ymax = 1.2,
  title = '$f(\\theta) = (\\sin\\theta )^2 - \\frac{1}{4}$'
    )

show( solution_points + trig_plot)
(solution_points + trig_plot ).save('trig-quadratic-problem-1-plot.svg')
```

```python
f(theta) = (sin(theta))^2 - 5*sin(theta) - (1/sqrt(2))*sin(theta) +  (5)/(sqrt(2))

xt = [k * pi for k in range(-3,4)]
xt_lbl = ['$-3\pi$', '$-2\pi$', '$-\pi$', '', '$\pi$', '$2\pi$', '$3\pi$',]

solutions = [ pi/4 + 2*k*pi for k in range(-1,2)] + [ 3*pi/4 + 2*k*pi for k in range(-1,2)] 

solution_points = points([(theta, f(theta)) for theta in solutions ], 
                                                 color = 'blue', 
                                         pointsize=30,
                                        zorder= 100)

trig_plot = plot(f,
         ( theta,-3*pi,3*pi),
             ticks = [xt, [-2,2,4,6,8,10]],
             tick_formatter = [xt_lbl, ['$-2$','$2$','$4$','$6$','$8$','$10$']],
     thickness = 3,
     color = 'red',
     aspect_ratio = 1,
                 ymin = -2.2,
                   title = '$f(\\theta) = \\sin^2(\\theta) - 5\\sin(\\theta) - \\frac{1}{\\sqrt{2}}\\sin(\\theta) + \\frac{5}{\\sqrt{2}}$'
    )

show( solution_points + trig_plot)
(solution_points + trig_plot ).save('trig-quadratic-problem-2-plot.svg')
```
